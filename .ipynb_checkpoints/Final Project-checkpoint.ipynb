```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd   

from sklearn.decomposition import PCA
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

from sklearn.model_selection import GridSearchCV
```

# Data preprocessing

```python
headerList = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol', 'quality']

wineData = pd.read_csv('winequality-red.csv', header = 0, names = headerList, sep=";")
print(wineData.head(10))

#Summarative functions
wineData.dtypes
wineData.describe()
wineData.info()

#Plotting histogram of each variable
wineData.hist(alpha=0.5, figsize=(15, 10))
plt.tight_layout()
plt.show()

for h in headerList:
  wineData[h] = pd.to_numeric(wineData[h], errors='coerce')

#checking to see null values
print("\nChecking for null values: \n")
wineData.isna().sum()
#we see there are no NaN cells

#We see there are 4 + 2 = 6 NaN cells 
#wineData = wineData.fillna(0)

#Check for null instances
#print("\nChecking for null values after using fillna(): \n")
#wineData.isna().sum()
```

# Data analysis

```python
#PCA
df_pca = wineData.copy()
X_pca = df_pca.loc[:, 'fixed acidity':'alcohol']
y_pca = df_pca['quality']

X_pca.tail()

#Preprocessing in preparation for PCA: Standardizing the predictor variables
X_pca = StandardScaler().fit_transform(X_pca)

#Fit PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_pca)

X_pca.shape

plt.figure(figsize=(8,6))
plt.scatter(X_pca[:,0],X_pca[:,1],c=y_pca,cmap='rainbow')
plt.xlabel('First principal component')
plt.ylabel('Second Principal Component')
plt.title("Using PCA to Visualize Classes")
plt.show()
```

```python
# feature selection using chi-square
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

X_f = wineData.loc[:, 'fixed acidity':'alcohol']
y_f = wineData['quality']

#apply SelectKBest class to extract top best features
bestFeatures = SelectKBest(score_func=chi2, k=11)
bestFeaturesFit = bestFeatures.fit(X_f,y_f)
dfscores = pd.DataFrame(bestFeaturesFit.scores_)  #Store predictor scores in a column 
dfcolumns = pd.DataFrame(X_f.columns)  #Store predictor variable names in a column

# #concatenate scores with predictor names
predScores = pd.concat([dfcolumns,dfscores],axis=1)
predScores.columns = ['Predictor','Score']   #naming the dataframe columns
print(predScores.nlargest(11,'Score'))       #print top (by score) 11 features

#Drop the bottom two features (smallest score)
selectedDf = wineData.drop('pH', 1)
selectedDf = selectedDf.drop('density', 1)

#Drop features with high correlation
selectedDf = selectedDf.drop('free sulfur dioxide', 1)
selectedDf = selectedDf.drop('fixed acidity', 1)

# selectedDf.dtypes
selX = selectedDf.loc[:, 'volatile acidity':'alcohol']
selY = selectedDf['quality']
```

```python
#feature selection using correlation matrix
import seaborn as sns

plt.figure(figsize=(12,10))
sns.heatmap(wineData.corr(),annot=True, cmap='coolwarm',fmt='.2f')

```

# modeling

```python
#Split data into training and test sets
X = wineData.loc[:, 'fixed acidity':'alcohol']
y = wineData['quality']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .3, random_state=10) #split the data
X_train.shape, y_train.shape, X_test.shape, y_test.shape

selX_train, selX_test, selY_train, selY_test = train_test_split(selX, selY, test_size = .3, random_state=10) #split the data
selX_train.shape, selY_train.shape, selX_test.shape, selY_test.shape

#Use sklearn.preprocessing.StandardScaler to standardize predictors 
scaledData = StandardScaler()
X_train = scaledData.fit_transform(X_train)
X_test = scaledData.transform(X_test)

selX_train = scaledData.fit_transform(selX_train)
selX_test = scaledData.transform(selX_test)
```

```python
#random forest
wineRF = RandomForestClassifier(n_estimators=200, max_depth=6, random_state=417)
wineRF.fit(X_train, y_train)

y_pred = wineRF.predict(X_test)

#Plot confusion matrix
#cm = confusion_matrix(y_test, y_pred)
#df_cm = pd.DataFrame(cm)
#sns.set(font_scale=1.2) # for label size
#sns.heatmap(df_cm, annot=True, annot_kws={"size": 16}) # font size
#plt.show()

#Print classification report
clas = classification_report(y_test, y_pred)
print(clas)

# Try with feature selected X - ended up performing poorer
wineRF = RandomForestClassifier(n_estimators=200, max_depth=6, random_state=417)
wineRF.fit(selX_train, selY_train)

selY_pred = wineRF.predict(selX_test)
# # confusion_matrix(y_test, y_pred)
clas = classification_report(selY_test, selY_pred)
print(clas)
```

```python
#svm
wineSVM = SVC(kernel = 'rbf')
wineSVM.fit(X_train, y_train)
y_pred = wineSVM.predict(X_test)

#Plot confusion matrix
# cm = confusion_matrix(y_test, y_pred)
# df_cm = pd.DataFrame(cm)
# sns.set(font_scale=1.2) # for label size
# sns.heatmap(df_cm, annot=True, annot_kws={"size": 16}) # font size
# plt.show()

#Print classification report for rbf kernel
clas1 = classification_report(y_test, y_pred)
print("rbf kernel classification report: \n",clas1)

#Print classification report for sigmoid kernel
wineSVM = SVC(kernel = 'sigmoid')
wineSVM.fit(X_train, y_train)
y_pred = wineSVM.predict(X_test)
clas2 = classification_report(y_test, y_pred)
print("sigmoid kernel classification report: \n", clas2)


#try with feature selected X
wineSVM = SVC(kernel = 'rbf')
wineSVM.fit(selX_train, selY_train)
selY_pred = wineSVM.predict(selX_test)

#Print classification report for rbf kernel
clas3 = classification_report(selY_test, selY_pred)
print("rbf kernel classification report: \n",clas3)

#Print classification report for sigmoid kernel
wineSVM = SVC(kernel = 'sigmoid')
wineSVM.fit(selX_train, selY_train)
selY_pred = wineSVM.predict(selX_test)
clas4 = classification_report(selY_test, selY_pred)
print("sigmoid kernel classification report: \n", clas4)
```

```python
#logistic regression
wineLR = LogisticRegression()
wineLR.fit(X_train, y_train)

# Predicting the Test set results
y_pred = wineLR.predict(X_test)

# cm_test = confusion_matrix(y_test, y_pred)
# df_cm = pd.DataFrame(cm)
# sns.set(font_scale=1.2) # for label size
# sns.heatmap(df_cm, annot=True, annot_kws={"size": 16}) # font size
# plt.show()

y_pred_train = wineLR.predict(X_train)
cm_train = confusion_matrix(y_pred_train, y_train)

print()
print('Accuracy for training set for Logistic Regression = {}'.format((cm_train[0][0] + cm_train[1][1])/len(y_train)))
print('Accuracy for test set for Logistic Regression = {}'.format((cm_test[0][0] + cm_test[1][1])/len(y_test)))
print()

clas = classification_report(y_pred, y_test)
print(clas)
```

# hyperparameter tuning

```python
#Create a default random forest object
wine_RFC = RandomForestClassifier(random_state=417)

#create a list of parameters you want to tune
param_grid_RFC = { 
    'n_estimators': [350,400, 450],
    'max_depth': [8,9,10]
}

#fit the model using grid search
CV_rfc = GridSearchCV(estimator=wine_RFC, param_grid=param_grid_RFC, cv= 10)
CV_rfc.fit(X_train, y_train)

#print the result of best hyperparameters
print(CV_rfc.best_params_)
```

```python
wine_SVM = SVC()
param_grid_svm = {
    'kernel': ['rbf','sigmoid','poly'], 
    'C': [1, 10, 100, 1000]
}

CV_svm = GridSearchCV(estimator=wine_SVM, param_grid=param_grid_svm, cv= 10)
CV_svm.fit(X_train, y_train)

#print the result of best hyperparameters
print(CV_svm.best_params_)
           
```

```python
wine_LR = LogisticRegression()
param_grid_lr = {
    'penalty' : ['l1', 'l2'],
    'solver':['lbfgs', 'liblinear','saga']
}

CV_lr = GridSearchCV(estimator=wine_LR, param_grid=param_grid_lr, cv= 10)
CV_lr.fit(X_train, y_train)

#print the result of best hyperparameters
print(CV_lr.best_params_)
```

# testing final model

```python
#using the best parameter we got from the grid search
wineRF_final = RandomForestClassifier(n_estimators=100, max_depth=8, random_state=417)
wineRF_final.fit(X_train, y_train)
y_pred = wineRF_final.predict(X_test)

#print out the accuracy of final model
print(wineRF_final.score(X_test,y_test))

#print out the classification report
clas = classification_report(y_pred, y_test)
print(clas)
```

```python
wineSVM_final = SVC(kernel = 'sigmoid', C = 1)
wineSVM_final.fit(X_train, y_train)
y_pred = wineSVM_final.predict(X_test)

print(wineSVM_final.score(X_test,y_test))

clas = classification_report(y_pred, y_test)
print(clas)
```

```python
wineLR_final = LogisticRegression(penalty='l1', solver= 'liblinear')
wineLR_final.fit(X_train, y_train)
y_pred = wineLR_final.predict(X_test)

print(wineLR_final.score(X_test,y_test))

clas = classification_report(y_pred, y_test)
print(clas)
```
